```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
# Spotify DEV portal id e pwd

CLIENT_ID = "***"
CLIENT_SECRET = "***"
```

```python
# spotify API connect

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=CLIENT_ID,
                                                           client_secret=CLIENT_SECRET))
```

```python
# countries list to iterate when gathering toplists

paises = ['AD', 'AR', 'AU', 'AT', 'BE', 'BO', 'BR', 'BG', 'CA', 'CL', 'CO', 'CR', 'CY',
          'CZ', 'DK', 'DO', 'EC', 'SV', 'EE', 'FI', 'FR', 'DE', 'GR', 'GT', 'HN', 'HK',
          'HU', 'IS', 'ID', 'IE', 'IT', 'JP', 'LV', 'LI', 'LT', 'LU', 'MY', 'MT', 'MX',
          'MC', 'NL', 'NZ', 'NI', 'NO', 'PA', 'PY', 'PE', 'PH', 'PL', 'PT', 'SG', 'ES',
          'SK', 'SE', 'CH', 'TW', 'TR', 'GB', 'US', 'UY']

# empty dataframe to add toplists' params

pl = pd.DataFrame(columns=["name", "uri", "country"])

# aux variables

tamanhos = dict()
dicio = {"name":[], "uri":[], "country":[]}
```

```python
# iterate through countries list requesting all the playlists from the "toplists" categories

for i in range(0,len(paises), 1):
    
    # get all the playlists from "toplists" categories from each country
    
    retorno = sp.category_playlists(category_id="toplists", country=paises[i], limit=50)
    tam = len(retorno["playlists"]["items"])
    tamanhos[paises[i]] = tam
    dicio = {"name":[],"uri":[],"country":[]}

    #for each playlist, store its name, uri (spotify ID) and origin country
    
    for x in range(0, tam, 1):
        
        dicio["name"].append(retorno["playlists"]["items"][x]["name"])
        dicio["uri"].append(retorno["playlists"]["items"][x]["uri"])
        dicio["country"].append(paises[i])
    
    # add stored data to a dataframe
    
    aux = pd.DataFrame(dicio, columns=["name", "uri", "country"])
    pl = pd.concat([pl, aux], ignore_index=True)    
```

```python
# world number of toplists playlists

sum(tamanhos.values())
```

```python
# dataframe containing world toplists playlists

pl
```

```python
# filter only playlists with "Top 50" on the title and dropping duplicates (by name and ID)

df = pl[pl["name"].str.contains("Top 50")]
df = df.drop_duplicates(subset=["name","uri"])
df.reset_index(inplace=True, drop=True)
```
